```python
import pandas as pd
import datetime
import numpy as np
```

```python
df = pd.read_csv('../data/preprocessed_ipp_factors.csv')

df.date = pd.to_datetime(df.date).dt.date
```

```python
r = pd.read_csv('../data/ключевая ставка.csv', index_col='Unnamed: 0')

r.date = pd.to_datetime(r.date, format="%Y-%m-%d").dt.date
r['month-date'] = r.date.apply(lambda date: f'{date.year}-{date.month:02}')
```

```python
exchange_rates = r.groupby('month-date', as_index=False)\
                    .agg({'rate': 'mean'})
```
